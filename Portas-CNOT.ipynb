# Portas CNOT
-Muda o estado do segundo q-bit, só e somente se ele for 1


-Ele é um q-bit que modula algo, a porta de cima controla a de baixo

```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(2,2)

qc.cx(0,1)#Porta CNOT (Primeira entrada[Qbit Controlador] // Segunda entrada[Q-bit Target])
qc.draw('mpl')
```

```python
qc = QuantumCircuit(2,2)

qc.x(0)
qc.cx(0,1)
qc.draw('mpl')
```

```python
qc.measure_all([0,1], [0,1])
qc.draw('mpl')
```

```python
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
backend = AerSimulator()
job = backend.run(qc, shot = 1024)

result = job.result()
count= result.get_counts()

plot_histogram(count)
```
